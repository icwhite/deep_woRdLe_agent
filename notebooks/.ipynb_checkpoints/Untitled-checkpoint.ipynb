```python
# Imports 
from utils.environments.wordle import WordleSimple
import numpy as np
```

```python
# Get wordle words
wordle_full =  [word.replace('\n', '').replace(',', '') for word in list(open('../scripts/wordle_words.txt', 'r'))]
wordle_100 = [word.replace('\n', '').replace(',', '') for word in list(open('../scripts/wordle_subset.txt', 'r'))]
np.random.seed(88)
wordle_500 = np.random.choice(wordle_full, 500)
```

```python
env = WordleSimple(valid_words = wordle_full, keep_answer_on_reset = False, logdir = None)
```

```python
actions = []
episode_returns = []
rewards = []
obses = []
episode_starts = []
n_episodes = 10_000
timestep = 0
for i in range(n_episodes): 
    episode_starts.append(timestep)
    
    obs = env.reset()
    done = False
    episode_return = 0
    while not done: 
        guess = 'audio' if env.guess_count == 0 else np.random.choice(env.possible_words)
        action = env.valid_words.index(guess)
        obs, reward, done, info = env.step(action)
        
        episode_return += reward
        timestep += 1
        
        # Append stuff 
        actions.append(action)
        rewards.append(reward)
        obses.append(obses)
        
    episode_returns.append(episode_return)
```

```python
import matplotlib.pyplot as plt
```

```python
d = {'action': actions, 
     'episode_returns': episode_returns, 
     'rewards': rewards, 
     'obs': obses, 
     'episode_starts': episode_starts}
```

```python
np.savez('test_expert_10k', d)
```

```python
import numpy as np
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.ppo import MlpPolicy

from imitation.algorithms import bc
from imitation.data import rollout
from imitation.data.wrappers import RolloutInfoWrapper
```

```python
expert_policy = np.load('test_expert_10k.npz')
```

```python
rng = np.random.default_rng(0)
env = WordleSimple(valid_words = wordle_full, keep_answer_on_reset = False, logdir = 'data')

rollouts = rollout.rollout(
    expert_policy,
    DummyVecEnv([lambda: RolloutInfoWrapper(env)]),
    rollout.make_sample_until(min_timesteps=None, min_episodes=50),
    rng=rng,
)
transitions = rollout.flatten_trajectories(rollouts)

bc_trainer = bc.BC(
    observation_space=env.observation_space,
    action_space=env.action_space,
    demonstrations=transitions,
    rng=rng,
)
bc_trainer.train(n_epochs=1)
reward, _ = evaluate_policy(bc_trainer.policy, env, 10)
print("Reward:", reward)
```
