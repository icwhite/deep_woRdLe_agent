```python
# Imports 
from tensorflow.python.summary.summary_iterator import summary_iterator
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
summary_iterator('data/stable_baseline_ppo_elim_reward_wordle_words_04-12-2022_17-56-08')
```

### Baseline

```python
# Grab Data 
path = 'data/stable_baseline_ppo_elim_reward_wordle_words_04-12-2022_17-56-08/PPO_1'
path += '/' + os.listdir(path)[0]
ep_len_list, ep_rew_list = [],[]

value_loss, policy_loss, alpha_loss = [],[],[]
for e in summary_iterator(path): 
    for v in e.summary.value: 
        if v.tag == 'rollout/ep_len_mean': 
            ep_len_list.append(v.simple_value)
        elif v.tag == 'rollout/ep_rew_mean': 
            ep_rew_list.append(v.simple_value)
        elif v.tag == 'train/value_loss': 
            value_loss.append(v.simple_value)
        elif v.tag == 'train/policy_loss': 
            policy_loss.append(v.simple_value)
        elif v.tag == 'train/entropy_loss': 
            entropy_loss.append(v.simple_value)
```

```python
# Episode Length 
fig, ax = plt.subplots(figsize = (8,4))
plt.subplots_adjust(bottom=0.25)
plt.plot(ep_len_list)
plt.xlabel('Iteration Number', labelpad = 5)
plt.ylabel('Average Episode Length', labelpad = 10)
plt.title('Baseline PPO Average Episode Length', font = 'serif', fontweight = 'bold', pad = 10)
fig.savefig('report_plots/ppo_baseline_ep_len.png')
```

```python
# Episode Length 
fig, ax = plt.subplots(figsize = (8,4))
plt.subplots_adjust(bottom=0.25)
plt.plot(ep_rew_list)
plt.xlabel('Iteration Number', labelpad = 5)
plt.ylabel('Average Episode Reward', labelpad = 10)
plt.title('Baseline PPO Average Episode Reward', font = 'serif', fontweight = 'bold', pad = 10)
fig.savefig('report_plots/ppo_baseline_reward.png')
```

```python
# Losses
fig, ax = plt.subplots(figsize = (8,4))
plt.subplots_adjust(bottom=0.15)

plt.plot(entropy_loss)

plt.xlabel('Iteration Number', labelpad = 5)
plt.ylabel('Entropy Loss', labelpad = 10)
plt.title('Baseline PPO Entropy Loss', font = 'serif', fontweight = 'bold', pad = 10)
fig.savefig('report_plots/ppo_baseline_entropy_loss.png')
```
